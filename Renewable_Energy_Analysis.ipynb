Step 1: Load the Data

```python
import pandas as pd

df = pd.read_csv("/renewable_energy.csv")
print(df[['Country', '%Renewable']])
```

Step 2: Filter the Data

```python
high_renewable = df[df["%Renewable"] > 50]
high_renewable
```

 Step 3: Create a Set

```python
high_set = set(high_renewable["Country"])
print("Countries with >50% renewable energy:", high_set)
print("Cardinality:", len(high_set))
```

Step 4: Create a Dictionary

```python
emissions_dict = dict(zip(df["Country"], df["CO2_Emissions"]))
print(emissions_dict["Brazil"])
```

Step 5: Bar Chart – CO₂ Emissions

```python
import matplotlib.pyplot as plt

plt.bar(emissions_dict.keys(), emissions_dict.values(), color="green")#add color in bar plot
plt.xticks(rotation=45)
plt.title("CO₂ Emissions per Capita by Country")
plt.ylabel("Tonnes")
plt.grid(True)

plt.show()
```

Step 6: Scatterplot – Population vs Renewable

```python
plt.scatter(df["Population"], df["%Renewable"])

for i in range(len(df)):
    plt.text(df["Population"][i], df["%Renewable"][i], df["Country"][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

Challenge: Plot only countries in the Asia region.

```python

asia_df = df[df["Region"] == "Asia"]

plt.scatter(asia_df["Population"], asia_df["%Renewable"])

for i in range(len(asia_df)):
    plt.text(asia_df["Population"].iloc[i], asia_df["%Renewable"].iloc[i], asia_df["Country"].iloc[i])

plt.title("Population vs % Renewable Energy Use (Asia Region)")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

Step 7: Save a Filtered Subset

```python
subset = df[df["%Renewable"] > 60]
subset.to_csv("high_renewable_countries.csv", index=False)
```
